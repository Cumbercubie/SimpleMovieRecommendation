```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class CF(object):
  #dist_function: measure similarity between 2 vectors, default is cosine
  #uuCF shows whether we are using user-user CF (1) or item-item CF (0)
  def __init__(self, y_data,k,dist_function = cosine_similarity, uuCF=1):
    self.uuCF = uuCF
    self.y_data = y_data if uuCF else y_data[:,[1,0,2]]
    self.k=k #number of neighborhoods
    self.dist_function = dist_function
    self.ybar_data = None
    self.n_users = int(np.max(self.y_data[:,0])) + 1
    self.n_items = int(np.max(self.y_data[:,1])) + 1 


  def add(self,new_data):
    self.y_data = np.concatenate((self.y_data,new_data),axis=0)


  def normalize_y(self):
    users = self.y_data[:,0]
    self.ybar_data = self.y_data.copy()
    self.mu = np.zeros((self.n_users,))
    for n in range(self.n_users):
      ids=np.where(users==n)[0].astype(np.int32)
      item_ids = self.y_data[ids,1]
      ratings = self.y_data[ids,2]
      m=np.mean(ratings)
      # if np.isnan(m):
      #   m=0
      self.ybar_data[ids,2] = ratings-self.mu[n]
    self.ybar = sparse.coo_matrix((self.ybar_data[:,2],(self.ybar_data[:,1],self.ybar_data[:,0])),(self.n_items,self.n_users))
    self.ybar=self.ybar.tocsr()

  def similarity(self):
    self.S=self.dist_function(self.ybar.T,self.ybar.T)

  def refresh(self):
      """   recalculate normalize & similarity matrix  """
      self.normalize_y()
      self.similarity() 
      
  def fit(self):
      self.refresh()
    
  def __pred(self,u,i,normalized=1):

    ids=np.where(self.y_data[:,1]==i)[0].astype(np.int32)
    users_rated_i = (self.y_data[ids,0]).astype(np.int32)
    sim = self.S[u,users_rated_i]
    a = np.argsort(sim)[-self.k]
    nearest_s = sim[a]
    r = self.ybar[i, users_rated_i[a]]
    if normalized:
            return (r*nearest_s)[0]/np.abs(nearest_s).sum()
    return (r*nearest_s)[0]/np.abs(nearest_s).sum() + self.mu[n]
  
  def pred(self,u,i,normalized=1):
    if self.uuCF: return self.__pred(u,i,normalized)
    return self.__pred(i,u,normalized)

  
  def recommend(self,u,normalized = 1):
    ids = np.where(self.y_data[:0]==u)[0]
    items_rated_by_u = self.y_data[ids,1].tolist()
    recommended_items = []
    for i in range(self.n_items):
      if i not in items_rated_by_u:
        rating = self.__pred(u,i)
        if rating > 0:
          recommended_items.append(i)
    return recommended_items
  
  def print_recommendation(self):
    print('Recommendation:')
    for u in range(self.n_users):
      recommended_items = self.recommend(u)
      if self.uuCF:
        print('\t Recommend item(s):', recommended_items, 'to user',u)
      else:
        print('\t Recommend item(s):',u,'to user(s)',recommended_items)
```

```python
def testWithMovieLens_UUCF(rate_train,rate_test):
    rs = CF(rate_train, k = 30, uuCF = 1)
    rs.fit()

    n_tests = rate_test.shape[0]
    SE = 0 # squared error
    for n in range(n_tests):
        pred = rs.pred(rate_test[n, 0], rate_test[n, 1], normalized = 0)
        # ** means ^ (pred-rate_test[n,2])^2 
        SE += (pred - rate_test[n, 2])**2 

    RMSE = np.sqrt(SE/n_tests)
    print('User-user CF, RMSE =', RMSE)

def testWithMovieLens_IICF(rate_train,rate_test):
    rs = CF(rate_train, k = 30, uuCF = 0)
    rs.fit()

    n_tests = rate_test.shape[0]
    SE = 0 # squared error
    for n in range(n_tests):
        pred = rs.pred(rate_test[n, 0], rate_test[n, 1], normalized = 0)
        SE += (pred - rate_test[n, 2])**2 

    RMSE = np.sqrt(SE/n_tests)
    print ('Item-item CF, RMSE =', RMSE)
```

```python
ratings = pd.read_csv('ratings.dat',sep='::',names=['id','movieId','score','unix_timestamps'],encoding='utf-8')
ratings.shape
```

***Import file chứa thông tin phim đã qua xử lý***


```python
movies = pd.read_csv('df.csv',sep=',')
movies.shape
```

***Loại bỏ những phim không có thông tin chi tiết***

```python
movies = movies[movies['cast']=='None'].dropna(axis=1)
movies.shape
```

***Loại bỏ những rating có chứa các phim đã bị loại***


```python
hasinfo_movies = []
rated_movies = ratings['movieId'].value_counts()
rated_movies_id = list(ratings['movieId'].value_counts().index)
for id in rated_movies_id:
  if id not in movies['movieId']:
    hasinfo_movies.append(id)
ratings = ratings.drop(hasinfo_movies)
ratings.shape
```

***Chia tập test, train và xử lý***

```python
from sklearn.model_selection import train_test_split

rate_train, rate_test = train_test_split(ratings, test_size=0.33, random_state=42)

rate_train = rate_train.to_numpy()
rate_test = rate_test.to_numpy()

rate_train[:, :2] -= 1
rate_test[:, :2] -= 1
```

```python
testWithMovieLens_IICF(rate_train,rate_test)
```

***Thống kê lại thông tin các phim đã được rate và id của chúng***

```python
rated_movies = ratings['movieId'].value_counts()
rated_movies_id = list(ratings['movieId'].value_counts().index)
```

***Thống kê 10 phim được rate nhiều nhất***

```python
top_10 = list(hasinfo_movies[:10])
top_10_rates = hasinfo_movies[:10]
top_10_titles = [movies[movies['movieId']==x]['title'].values[0].split("(")[0] for x in top_10]
# for x in top_10:
#   title = movies[movies['movieId']==x]['title'].values
#   print(title)
top_10_df = pd.DataFrame({"movieId":top_10,"title":top_10_titles,"rate_counts":top_10_rates})
top_10_rates
```

```python
top_10_df
```

```python
fig_dims = (20, 9)
fig, ax = plt.subplots(figsize=fig_dims)
ax.legend(loc='best',fontsize=50)
some_plot = sns.barplot(y='title',x='rate_counts',data=top_10_df,ax=ax,orient='h')
# top_10_df
plt.show()
```

***Thống kê 10 phim được đánh giá cao nhất***

```python
avg_rates = []
rate_max = 0
rate_max_id = ""
for movieId in rated_movies_id:
    rates = ratings[ratings['movieId']==movieId]['score'].values
    rate_counts = len(rates)
    if rate_counts == 0:
      print(movieId)
    else:
      rate_sum = sum(rates)
      avg = rate_sum/rate_counts
      if avg > rate_max: 
        rate_max = avg
        rate_max_id = movieId 
      avg_rates.append({'Id':movieId,'score':avg})
    # print(str(avg) + " "+ str(movies[movies['movieId']==movieId]['title'].values[0]))
# print(str(rate_max) +" "+str(rate_max_id))
sorted_scores = sorted(avg_rates, key = lambda i: i['score'],reverse=True)
```

```python
highest_10 = sorted_scores[:10]
highest_10_df = pd.DataFrame(highest_10)
highest_10_titles = [movies[movies['movieId']==x['Id']]['title'].values[0].split("(")[0] for x in highest_10]
highest_10_df['title'] = highest_10_titles 
```
