```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tmdb_df = pd.read_csv('TMDB.txt',sep=',')
tmdb_df.columns
```

```python
ml_25 = pd.read_csv('./ml-25m/movies.csv')
ml_25.columns
```

```python
ml_25_movies = ml_25[['movieId','title']]
```

```python
import re
ml_25_title = list(ml_25_movies['title'].values)
# ml_25_name = [x[0].strip() for x in ml_25_title]
ml_year=[]
for index,titles in enumerate(ml_25_title):
#         print(titles)
        year = re.findall(r'\(([1-3][0-9]{3})\)',titles) 
        ml_year.append(year[0]) if len(year)>0 else ml_year.append(None)
        titles = titles.split('(')[0].strip()
        ml_25.at[index,'title']=titles
ml_year
df_year = pd.DataFrame(ml_year,columns=['Year'])
ml_25['year'] = df_year
ml_25
```

```python
tmdb_movies = tmdb_df[['id','original_title']]
tmdb_title = list(tmdb_movies['original_title'])
tmdb_title
```

***Loại bỏ các phim trùng bằng cách bỏ phim có thời gian ra mắt lâu hơn, do các phim cũ thường sẽ ít đánh giá hơn***

```python
freq_dict = {}
for title in tmdb_df['original_title']:
    if title in freq_dict: #means there's another movie already
#       get years of duplicated movies 
        years_and_id = tmdb_df.loc[tmdb_df['original_title']==title][{'release_date','id'}].values
        id_min_year = min([movie[0] for movie in years_and_id])
#         print(id_min_year)
        tmdb_df.drop(tmdb_df[tmdb_df['id']==id_min_year].index,axis=0,inplace=True)
        freq_dict[title] = years_and_id
    else:
        freq_dict[title] = []
# freq_dict
```

***Kiểm tra lại sau khi loại bỏ các phim trùng***

```python
freq_dict = {}
for title in tmdb_df['original_title']:
    if title in freq_dict: #means there's another movie already
        freq_dict[title] +=1
    else:
        freq_dict[title] = 0
freq_dict
```

```python
def sim_rate(df1,df2):
    similar=[]
    for title in df1:
        if str(title) in df2:
            similar.append(1)
        else:
            similar.append(0)
    return (similar.count(1),similar.count(0))
```
